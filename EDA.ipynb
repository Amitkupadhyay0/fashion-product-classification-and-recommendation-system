# EDA of Fashion data set

The data has been divided into two part : Train , Test

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
## Read Train data CSV file
df_train=pd.read_csv(r"C:\Users\amitu\Music\fashion_dataset\train_data.csv")
## Read Test data
df_test=pd.read_csv(r"C:\Users\amitu\Music\fashion_dataset\test_data.csv")
```

```python
df_train.head(5)
```

The data frame have
1. id --> Unique number of each image
2. article type --> category to which image belong 
3. baseColour --> colour of product
4. gender --> gender 
5. brandName

### 1. Total data

```python
len(df_train) + len(df_test)
```

There are total 22307 fashion product in this data

 ### 2. Count of articleType 

    1. In train data

```python
unique_article_counts = df_train['articleType'].value_counts()
# Sort the counts
unique_article_counts = unique_article_counts.sort_values(ascending=False)
## Plot of count
num_colors = len(unique_article_counts)
colors = plt.cm.viridis(np.linspace(0, 1, num_colors))  # You can choose a different colormap
# histogram
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
unique_article_counts.plot(kind='bar', color=colors)
plt.title("Count of Unique Article Types")
plt.xlabel("Article Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

     2. In test data

```python
unique_article_counts = df_test['articleType'].value_counts()
# Sort the counts
unique_article_counts = unique_article_counts.sort_values(ascending=False)
## Plot of count
num_colors = len(unique_article_counts)
colors = plt.cm.viridis(np.linspace(0, 1, num_colors))  # You can choose a different colormap
# histogram
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
unique_article_counts.plot(kind='bar', color=colors)
plt.title("Count of Unique Article Types")
plt.xlabel("Article Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

Test data is almost balance
